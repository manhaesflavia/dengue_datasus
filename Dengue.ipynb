```python
#! pip install geopandas
import pandas as pd

# ------ Visualizar todos os arquivo da pasta
from os import chdir, getcwd, listdir
from os.path import isfile
# importing re for regular expressions
import re
#!pip install unidecode
from unidecode import unidecode

# --------- Diretório com todos os arquivos no formato csv
path = r"H:\.shortcut-targets-by-id\1_JLmrBaruoIgBXkCr76gEhHdz8pd_J-A\DADOS\Doenças\02_Dengue"

# -------- Estações convencionais (INMET)

# ---- diretorio dos arquivos
chdir(path)
print(getcwd())

# --------- Criando uma lista para visualizar todos os arquivos
todosOsArquivos = []
for c in listdir():
    if isfile(c):
        todosOsArquivos.append(c)

# ---- criar um dataframe com a listagem de todos os arquivos
df_arquivosDiretorio = pd.DataFrame()

# ---- criando uma coluna com o nomes dos arquivos
df_arquivosDiretorio["todosOsArquivos"]=todosOsArquivos

# ------ Aplicando um filtro de acordo com o nome do arquivo
df_arquivosDiretorio = df_arquivosDiretorio[df_arquivosDiretorio["todosOsArquivos"].str.contains("dengue")].reset_index(drop=True)

# ----- Visualizando todos os arquivos no diretorio
print(df_arquivosDiretorio)

# ------ Criando um df vazio para colocar os dados de dengue
dadosDengue = pd.DataFrame()
# ------ Criando um laço para ir lendo cada arquivo do diretorio
for i in range(len(df_arquivosDiretorio['todosOsArquivos'])):
    print(path + "\\" + df_arquivosDiretorio['todosOsArquivos'][i])
    # ----- Concatenando todas as planilhas para os das estações
    
    # Aqui fazendo com que ele pule as 3 primeiras linhas que sao o cabeçalho
    tabela = pd.read_csv(path + "\\" + df_arquivosDiretorio['todosOsArquivos'][i], encoding = 'iso8859_14',
            sep=";", skiprows = 3)
    
    # ------- Pulando apenas as duas primeiras para poder obter o periodo que se trata o conjunto de dados
    periodo  = pd.read_csv(path + "\\" + df_arquivosDiretorio['todosOsArquivos'][i], encoding = 'iso8859_14',
            sep=";", skiprows = 2).columns[0].split(":")[1]
    
    tabela = tabela[tabela['Município de notificação'].str.startswith('26')]
    tabela['Período'] = periodo
    
    dadosDengue = pd.concat([dadosDengue, tabela])
    
df_Dengue = pd.DataFrame()
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
       'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

for i in range(len(meses)):
    Dengue = dadosDengue[['Município de notificação', meses[i], 'Período']]
    Dengue['Mês'] = meses[i]
    Dengue.columns = ['Município de notificação', 'Nº de casos', 'Período', "Mês"]
    
    df_Dengue = pd.concat([df_Dengue, Dengue])

df_Dengue['Nº de casos'] = df_Dengue['Nº de casos'].str.replace('-',"0")    

df_Dengue = df_Dengue.reset_index(drop=True)
Municipio = []
Geocodigo = []
for i in range(df_Dengue.shape[0]):
    
    Geocodigo.append(df_Dengue['Município de notificação'][i].split(" ", 1)[0])
    Municipio.append(df_Dengue['Município de notificação'][i].split(" ", 1)[1])
    
df_Dengue['Município de notificação'] = Municipio
df_Dengue['Geocodigo'] = Geocodigo

# ---- Criando um ID para a linha
df_Dengue = df_Dengue.reset_index()
df_Dengue.columns = ['ID', 'NM_Municipio', 'N_Casos', 'Ano', 'Mes',
       'Geocodigo']

display(df_Dengue)
```
